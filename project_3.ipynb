### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score, accuracy_score

knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(x_train_glove300, y_train_glove300)
predictions = list(knn.predict(x_test_glove300))
KNN_dev_predictions = list(knn.predict(x_dev_glove300))
print("KNN Accuracy:",knn.score(x_dev_glove300,y_dev_glove300))
print("KNN F1-score:",f1_score(y_dev_glove300,KNN_dev_predictions, average='macro'))
```
